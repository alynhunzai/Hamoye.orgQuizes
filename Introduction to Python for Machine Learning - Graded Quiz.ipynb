```python
# importing libraries
import numpy as np
import pandas as pd
```

```python
# Read the data into 'df'
df = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding='latin-1')
df
```

```python
# Groupby operation on 'Item' columns 
df[['Item' , 'Y2015', 'Y2018']].groupby('Item').sum()
```

```python
# Correlation of 'Element Code' with all of the years' cols 
df[['Element Code', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].corr()
```

```python
# Groupby operation on 'Element' col for 'Stock Variation'
df.groupby('Element')[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].sum()
```

```python
# Mean and Standard Deviation for year 2017 limited to 2 decimal places
y2017_mean = round(df.Y2017.mean(), 2)
y2017_std = round(df.Y2017.std(), 2)
print('Mean of 2017: ', y2017_mean)
print('Std of 2017: ', y2017_std)
```

```python
# Unique countries in the dataset
df.Area.nunique()
```

```python
# Groupby on 'Area' and 'Y2017' to find highest sum
df[['Area', 'Y2017']].groupby('Area').sum().sort_values('Y2017', ascending=False)
```

```python
# Protein supply in year 2015 in Madagascar
df[df['Area']=='Madagascar'].groupby(['Element'])['Y2015'].sum()
```

```python
# Groupby on 'Area' and 'Y2017' to find 7th lowest sum in 2017
df[['Area', 'Y2017']].groupby(['Area']).sum().sort_values('Y2017')
```

```python
# Groupby operation on 'Element' col for total sum of 'Processing' in 2017
df.groupby('Element')[['Y2017']].sum()
```

```python
# Missing values in 'Y2014' col
total_missing = df[['Y2014']].isnull().sum().sum()
pct_missing = round((total_missing / df.shape[0]) * 100, 3)
print(f'Total missing values in 2014: {total_missing}')
print(f'Percent missing values in 2014: {pct_missing}%')
```
